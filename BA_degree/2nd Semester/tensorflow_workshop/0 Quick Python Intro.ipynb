## The loops
So how do we actually use it for something?

```
# First and FORemost, the FOR loop
max_int = 10
for i in range(max_int):
    print(i)
```

```
## Now the smart part of python for loops
students_name = ["Annabelle", "Tony", "Neo", "Heisenberg", "Jon Snow"]
for student in students_name:
    print(student)
```

```
# Now the while loops

is_it_below_5 = True
current_value = 0
while is_it_below_5:
    current_value += 1 # the same as ++ in e.g. C++
    print(f"Current value is {current_value}")

    ## Here we have the if statement
    if current_value >= 5:
        print("Terminating loop")
        is_it_below_5 = False
    
```
