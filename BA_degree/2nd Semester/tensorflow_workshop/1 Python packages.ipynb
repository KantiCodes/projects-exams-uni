# Matplotlib
Plotting done right

```
import matplotlib.pyplot as plt # to make it easy for us :-) 
```

Lets define a function that returns some nice data

```
def f(t):
    # Notice we use the numpy module here for the exponential function
    # We could also use the built-in module "math" for math.exp
    return t**2*np.exp(-t**2)
```

Now lets define some points to calculate the function on

```
t = np.linspace(start=0, stop=4, num=100) # 100 points between 0 and 4
y = f(t)                  # compute all f values
```

Now lets plot it - it is easy

```
plt.plot(t,y) # t on x axis, and y on y axis
plt.savefig('first-plot.pdf') # produce PDF
plt.savefig('first-plot.png') # produce PNG
```

# Pandas 
Easy data handling of e.g. CSV files and more

```
import pandas as pd
```

Import a csv file to pandas dataframe

```
df = pd.read_csv('sample_data/california_housing_test.csv')
df
```

Now lets look at the median house value for the cities in California. 

```
m_house_value = df['median_house_value']
m_house_value.plot()
```

Okay, we do not get a large insight from this plot, let see if there are some corrolations between the datapoints.

```
df.plot(x='median_income', y=['median_house_value'], style='o')
```

Okay, so here there are some corrolations between median income and median house value.

What else should we try? 

```
df.plot(x='population', y=['median_house_value'], style='*')
```
